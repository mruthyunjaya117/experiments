```python
import PyPDF2
with open('PragmaticProgrammer.pdf', 'rb') as pdf_file:
    reader = PyPDF2.PdfFileReader(pdf_file)
    num_pages = reader.getNumPages()
    
    data = ""
    
    for page_no in range(num_pages):
        page = reader.getPage(page_no)
        data += page.extractText()
```

```python
with open('pragmatic.txt','w', encoding='utf-8') as txt_file:
    txt_file.write(data)
```

```python
from sklearn.model_selection import train_test_split

with open('pragmatic.txt', 'r', encoding='utf-8') as file:
    text = file.read()
    
    train_data, val_data = train_test_split(text, test_size=0.3, random_state=8)
    
with open('prag_train.txt','w', encoding='utf-8') as file:
    file.write(str(train_data))
    
with open('prag_val.txt','w', encoding='utf-8') as file:
    file.write(str(val_data))
```
